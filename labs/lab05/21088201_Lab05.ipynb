```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
telecom_data = pd.read_csv('./data/telecom_churn_clean.csv')

telecom_data.head()
```

```python
# 1. Binary classification
# There are two types of supervised learning â€” classification and regression. 
# Binary classification is used to predict a target variable that has only two labels, 
# typically represented numerically with a zero or a one. Your task is to examine the data 
# and choose which column could be the target variable for binary classification.

print("There are 3 columns that could be the target variable for binary classifications:")

binary_cols = [col for col in telecom_data.columns if sorted(telecom_data[col].unique()) == [0, 1]]
print("Columns suitable for binary classification: ", binary_cols)
```

```python
# The workflow of building a supervised learning model and making predictions.
# 2. Kháº£o sÃ¡t: fit vÃ  predict
from sklearn.neighbors import KNeighborsClassifier

X = telecom_data[['total_day_minutes', 'total_day_charge', 'total_eve_minutes', 'total_eve_charge', 'total_night_minutes', 'total_night_charge', 'total_intl_minutes', 'total_intl_charge']].values
y = telecom_data['churn'].values

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)

# Predict
X_new = np.array([[200, 20, 200, 20, 200, 20, 20, 2]])
prediction = knn.predict(X_new)
print("Prediction: ", prediction)
```

```python
# Using scikit-learn to fit a classifier
from sklearn.neighbors import KNeighborsClassifier

X = telecom_data[['total_day_charge', 'total_eve_charge']].values
y = telecom_data['churn']

knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X, y)
```

```python
# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21, stratify=y)

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```

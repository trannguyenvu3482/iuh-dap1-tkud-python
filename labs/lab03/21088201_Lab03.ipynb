# 1. Dataset: Spotify



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample 1000 rows from spotify_population, assigning to spotify_sample
spotify_population = pd.read_csv('./data/data_spotify.csv')
    
spotify_sample = spotify_population.sample(1000)
```

```python
# Calculate the mean duration in minutes from spotify_population using pandas
mean_duration_population = spotify_population['duration_ms'].mean() / 60000
print("Mean duration in minutes of pop: ", round(mean_duration_population, 4))

# Calculate the mean duration in minutes from spotify_sample using pandas
mean_duration_sample = spotify_sample['duration_ms'].mean() / 60000
print("Mean duration in minutes of sample: ", round(mean_duration_sample, 4))
```

# 2. Compare acousticness

```python
plt.subplot(2, 1, 1)
plt.hist(spotify_population['acousticness'], bins=100, alpha=0.5, edgecolor='black')
plt.title('Population acousticness')

plt.subplots_adjust(hspace=0.5)

plt.subplot(2, 1, 2)
plt.hist(spotify_sample['acousticness'], bins=100, alpha=0.5, edgecolor='black')
plt.title('Sample acousticness')

plt.show()

print("The acousticness values in the sample are not generalizable to the population.")
```

# 3. Vectors / 2D Array

```python
# Given a, b, m, n, with a < b, m and n > 0. Create a func that create a 2D array of shape (m, n) with random integers in [a, b)
def create_2D_array(a, b, m, n):
    return np.random.randint(a, b, size=(m, n))

# Test
print(create_2D_array(1, 10, 3, 4))
```

```python
# Given a is a random float, s is a random positive float, m and n are positive integers, create a func that returns a 2D array of shape (m, n) with the items of the array is being normal distributed with mean a and standard deviation s
def create_normal_distributed_array(a, s, m, n):
    return np.random.normal(a, s, (m, n))

# Test
print(create_normal_distributed_array(0, 1, 3, 4)) 
```

# 4. Generating random numbers

```python
# Generate 5k numbers from a uniform distribution, setting the params low to -3, high to 3
uniform_data = np.random.uniform(-3, 3, 5000)

# Generate 5k numbers from a normal distribution, setting the params loc to 5, scale to 2
norm_data = np.random.normal(5, 2, 5000)

# Plot the uniform_data with bins of width 0.25 from -3 to 3
plt.hist(uniform_data, bins=np.arange(-3, 3.25, 0.25), alpha=0.5, edgecolor='black')
```

```python
# Plot the norm_data with bins of width 0.5 from -2 to 13
plt.hist(norm_data, bins=np.arange(-2, 13.5, 0.5), alpha=0.5, edgecolor='black')
```

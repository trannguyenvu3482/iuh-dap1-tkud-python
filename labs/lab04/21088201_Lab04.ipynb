```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
attrition_pop = pd.read_csv('./data/data-HR-Employee-Attrition.csv')

attrition_pop.head()
```

```python
# Ex1: Sample 70 rows from attrition_pop using simple random sampling, setting the random seed to 18900217.
# Print the sample dataset, attrition_samp. What do you notice about the indices?
attrition_samp = attrition_pop.sample(70, random_state=18900217)
print(attrition_samp)
```

```python
# 2. Set the sample size to 70.
# Calculate the population size from attrition_pop.
attrition_samp = attrition_pop.sample(70, random_state=18900217)
attrition_pop_size = len(attrition_pop)
print("Population size: ", attrition_pop_size)

# Calculate the interval between the rows to be sampled.
# Systematically sample attrition_pop to get the rows of the population at each interval, 
# starting at 0; assign the rows to attrition_sys_samp.
interval = attrition_pop_size / 70
print("Interval: ", interval)
attrition_sys_samp = attrition_pop.iloc[::int(interval)]
```

```python
# 3.Get the proportion of employees by Education level from attrition_pop.
# Use proportional stratified sampling on attrition_pop to sample 40% of each Education group, setting the seed to 2022.
attrition_strat = attrition_pop.groupby('Education').apply(lambda x: x.sample(frac=0.4, random_state=2022))

# Get the proportion of employees by Education level from attrition_strat.
attrition_strat_prop = attrition_strat['Education'].value_counts(normalize=True)

attrition_strat_prop
```

```python
# 4. Create a list of unique JobRole values from attrition_pop, and assign to job_roles_pop
job_roles_pop = attrition_pop['JobRole'].unique()

# Randomly sample four JobRole values from job_roles_pop.
job_roles_samp = np.random.choice(job_roles_pop, 4, replace=False)

# Subset attrition_pop for the sampled job roles by filtering for rows where JobRole is in job_roles_samp.
attrition_samp = attrition_pop[attrition_pop['JobRole'].isin(job_roles_samp)]

# Remove any unused categories from JobRole.
attrition_samp['JobRole'] = pd.Categorical(attrition_samp['JobRole'], categories=job_roles_samp, ordered=False)

# For each job role in the filtered dataset, take a random sample of ten rows, setting the seed to 2022.
attrition_samp = attrition_samp.groupby('JobRole').apply(lambda x: x.sample(10, random_state=2022))

attrition_samp
```

## Ex 05 - 09

```python
# 5. Load the dataset and find the most common gender and birth country
nobel_pop = pd.read_csv('./data/nobel_laureates_data.csv')

# Find the top gender and top country
top_gender = nobel_pop['gender'].value_counts().idxmax()
top_country = nobel_pop['bornCountry'].value_counts().idxmax()

print("Top gender: ", top_gender)
print("Top born country: ", top_country)
```

```python
# 6. Find the decade that had highest proportion of US-born winners
nobel_pop['decade'] = (np.floor(nobel_pop['year'] / 10) * 10).astype(int)

max_decade_usa = nobel_pop[nobel_pop['bornCountry'] == 'USA']['decade'].value_counts().idxmax()
print("Max decade USA: ", max_decade_usa)
```

```python
# 7. Find the decade and category pair had the highest proportion of female laureates
nobel_pop_decade_category_size_max = nobel_pop[nobel_pop['gender'] == 'female'].groupby(['decade', 'category']).size().idxmax()

max_female_dict = {nobel_pop_decade_category_size_max[0]: nobel_pop_decade_category_size_max[1]}
max_female_dict
```

```python
# 8. Find the first woman to win a Nobel and the category
nobel_pop_first_woman = nobel_pop[nobel_pop['gender'] == 'female'].sort_values('year').iloc[0]

first_woman_name = nobel_pop_first_woman['fullName']
first_woman_category = nobel_pop_first_woman['category']

print("First woman to win a Nobel: ", first_woman_name)
print("Category: ", first_woman_category)
```

```python
# 9. Which individuals or organizations have won multiple Nobel Prizes throughout their lifetime?
multi_nobel_list = nobel_pop['fullName'].value_counts()[nobel_pop['fullName'].value_counts() > 1]

repeat_list = multi_nobel_list.index.tolist()
repeat_list
```

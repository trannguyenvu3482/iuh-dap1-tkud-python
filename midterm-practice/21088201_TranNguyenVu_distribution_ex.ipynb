# Distribution midterm practice

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
mu = 172
sigma = 5
```

```python
# a) Xác suất chọn ngẫu nhiên một bạn gọi là lùn (nhỏ hơn 162(cm))?
print("Xác xuất chọn ngẫu nhiên một bạn gọi là lùn: ", stats.norm.cdf(162, mu, sigma).round(4))
```

```python
# b) Giả sử, thanh niên nam được xem là cao lớn (> 182 (cm)). Hỏi xác suất chọn ngẫu nhiên một thanh 
# niên nam cao lớn ở Hàn Quốc là bao nhiêu?
print("Xác xuất chọn ngẫu nhiên một thanh niên nam cao lớn: ", (1 - stats.norm.cdf(182, mu, sigma)).round(4))
```

```python
# c) Chọn thanh niên nam có chiều cao gọi là tầm thước (trong khoảng 165-182(cm))? Hỏi xác suất chọn 
# người tầm thước là bao nhiêu
print("Xác xuất chọn ngẫu nhiên một thanh niên nam tầm thước: ", (stats.norm.cdf(182, mu, sigma) - stats.norm.cdf(165, mu, sigma)).round(4))
```

```python
# d) Cho biết 6% là thấp bé, vậy thế nào là chuẩn thấp bé của thanh niên nam ở Hàn Quốc (cm)?
print("Chuẩn thấp bé của thanh niên nam ở Hàn Quốc: ", stats.norm.ppf(0.06, mu, sigma).round(4))
```

```python
# e) Cho biết đến 9% cao lớn, vậy cao lớn có ngưỡng chiều cao là bao nhiêu (cm)? 
print("Cao lớn có ngưỡng chiều cao: ", stats.norm.ppf(0.91, mu, sigma).round(4))
```

```python
# f) Vẽ biểu đồ phân bố mô tả phân bố chuẩn trên 
sns.histplot(stats.norm.rvs(loc=mu, scale=sigma, size=10000), kde=True, bins=100)
plt.xlabel('Chiều cao')
plt.title('Phân bố chuẩn chiều cao')

plt.show()
```
